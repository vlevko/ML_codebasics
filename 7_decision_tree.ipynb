### Subject
Using the *titanic.csv* file with passengers dataset, specifically fields *Survived, Pclass, Sex, Age, Fare*, predict the survival rate based on specified features.

```python
import pandas as pd
import math
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df = df[['Survived','Pclass','Sex','Age','Fare']]
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df['Survived']
```

```python
le = LabelEncoder()
inputs['Sex'] = le.fit_transform(inputs['Sex'])
inputs.head(10)
```

```python
inputs['Age'] = inputs['Age'].fillna(math.floor(inputs['Age'].mean()))
inputs.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
